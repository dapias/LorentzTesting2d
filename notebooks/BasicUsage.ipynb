```julia
ρ = 0.46
bt = billiard_lorentz(ρ) #Billiard_table
p = randominside(bt) #Particle
plot_billiard(bt)
plot_particle(p)
tf = 10.0
xt, yt, vxt, vyt, t = construct(evolve!(p, bt, tf)...)
plot_billiard(bt, xt, yt)
plot_particle(p)
```
